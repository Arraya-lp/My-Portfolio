## Get data from public API

**Course** Python - Data Science Bootcamp

```python
## Get data about pokemon characters from public API
import requests
import time

pokemons = []

for i in range(5):
    url = f"https://pokeapi.co/api/v2/pokemon/{i+36}"
    resp = requests.get(url)
    if resp.status_code == 200:
        data = resp.json()

        pokemons_detail = {
            "Name": data["name"],
            "Weight": data["weight"],
            "Height": data["height"],
            "Abilities": [ability["ability"]["name"] for ability in data["abilities"]]
        }

        pokemons.append(pokemons_detail)

        time.sleep(1)


print(pokemons)
```

```python
import pandas as pd

pokemons = pd.DataFrame(pokemons)

pokemons
```
